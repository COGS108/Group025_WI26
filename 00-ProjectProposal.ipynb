## Research Question


Which non-clinical variables - including sex, race, ethnicity, BMI, insurance, income, physical activity, smoking, and drinking - are most predictive of diabetes among US adults in NHANES 2017-2018? How does this association change after controlling for basic socioeconomic factors (education, income, insurance)?



## Background and Prior Work

Diabetes is a chronic disease that occurs when blood glucose levels are too high. This is caused by either the pancreas not producing the hormone insulin or insulin receptors becoming unresponsive. Diabetes affects 1 in 10 Americans, with almost half of these cases remaining undiagnosed.

Several studies have attempted to utilize machine learning to predict diabetes incidence in patients by using clinical data both simple (blood pressure, BMI) to those more difficult to measure (adiposity, HbA1c, lipid profiles). These algorithms have achieved high accuracy.

Nevertheless, it is estimated than up to 25% of US adults skip annual healthcare checkups and up to 90% delay them. For these reasons, there are a number of at-risk patients who do not have up to date clinical data, let alone actual blood glucose tests. This project attempts to see if an accurate diabetes test could be developed using only lifestyle data that any person would know about themselves.

There is one study that has looked at this before called "Learning from the machine: is diabetes in adults predicted by lifestyle variables? A retrospective predictive modelling study of NHANES 2007â€“2018." This paper uses an XGBoost model and claims a promisingly high accuracy of prediction. However, this paper does little to interpret its results, such as quantifying which factors exactly are most predictive and concerns of equity (especially socioeconomic equity) could affect diabetes prevalence. This seems to us a critical gap in the literature that could give important insight into the development of diabetes and its eventual health burden.

## Hypothesis


Hypothesis: the individual clinical variables most associated with diabetes will be race, sex and BMI. After controlling for socioeconomic factors, the effects of BMI, sex, smoking and drinking will be more pronounced, revealing the effects the social determinants of health (like income, neighborhood, education) have on diabetes prevalence.

## Data

The data will be obtained from NHANES, which is a nationally representative CDC survey that provides public use data. Importantly, NHANES provides high-quality interviews, examinations and laboratory tests, enabling more qualitative data such as intensity of exercise (as opposed to just minutes per week) to be collected.

We will use the 2017-2018 data because this is the same dataset that the previous study on this topic used. This will allow us to compare our results with theirs to better establish that our findings are replicable and real.

## Ethics 

NHANES provides public-use, de-identified, self-reported data. For this reason, most of the typical ethical questions of data security and informed consent are not a concern.

We will take care to ensure that the self-reported data does not bias results.
