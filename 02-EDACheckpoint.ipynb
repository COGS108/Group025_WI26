```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data/01-interim/nhanes_diabetes.csv")

# biomarker-positive only (your analytic sample)
analytic = df[df["diabetes_biomarker"] == True].copy()

# outcome
analytic["undiagnosed"] = analytic["diagnosed"] == False
analytic["diagnosis_status"] = np.where(analytic["undiagnosed"], "Undiagnosed", "Diagnosed")

analytic.shape
```

```python
[c for c in analytic.columns if "WT" in c.upper()]
```

```python
W = "WTMEC2YR"
analytic[[W]].head()
```

```python
overall = analytic[["diagnosis_status", W]].dropna()

overall_prop = overall.groupby("diagnosis_status")[W].sum()
overall_prop = (overall_prop / overall_prop.sum()).reset_index()
overall_prop = overall_prop.rename(columns={W: "prop"})

overall_prop
```

```python
sns.barplot(data=overall_prop, x="diagnosis_status", y="prop")
plt.ylabel("Weighted Proportion")
plt.xlabel("")
plt.title("Weighted % Diagnosed vs Undiagnosed (Biomarker-Positive)")
plt.ylim(0, 1)
plt.show()
```

```python
gender = analytic[["RIAGENDR", "diagnosis_status", W]].dropna()

gender_prop = (
    gender.groupby(["RIAGENDR", "diagnosis_status"], as_index=False)[W].sum()
)

# within each gender, convert to proportions
gender_prop["prop"] = gender_prop[W] / gender_prop.groupby("RIAGENDR")[W].transform("sum")

gender_prop
```

```python
# optional: make labels readable
gender_prop["RIAGENDR"] = gender_prop["RIAGENDR"].replace({1: "Male", 2: "Female"})

sns.barplot(data=gender_prop, x="RIAGENDR", y="prop", hue="diagnosis_status")
plt.ylabel("Weighted Proportion within Gender")
plt.xlabel("Gender")
plt.title("Weighted Diagnosis Status by Gender")
plt.ylim(0, 1)
plt.legend(title="")
plt.show()
```

```python
race = analytic[["RIDRETH1", "diagnosis_status", W]].dropna()

race_prop = (
    race.groupby(["RIDRETH1", "diagnosis_status"], as_index=False)[W].sum()
)

# convert to within-group proportions
race_prop["prop"] = race_prop[W] / race_prop.groupby("RIDRETH1")[W].transform("sum")

race_prop
```

```python
# optional: make readable labels
race_prop["RIDRETH1"] = race_prop["RIDRETH1"].replace({
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    5: "Other"
})

sns.barplot(data=race_prop, x="RIDRETH1", y="prop", hue="diagnosis_status")

plt.ylabel("Weighted Proportion within Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.title("Weighted Diagnosis Status by Race/Ethnicity")
plt.ylim(0,1)
plt.xticks(rotation=30)
plt.legend(title="")

plt.show()
```

```python
def weighted_quantile(values, weights, q):
    v = np.asarray(values)
    w = np.asarray(weights)
    m = ~(np.isnan(v) | np.isnan(w))
    v, w = v[m], w[m]
    sorter = np.argsort(v)
    v, w = v[sorter], w[sorter]
    cw = np.cumsum(w)
    cutoff = q * cw[-1]
    return v[np.searchsorted(cw, cutoff)]
```

```python
age = analytic[["RIDAGEYR", "diagnosis_status", W]].dropna()

age_stats = age.groupby("diagnosis_status").apply(
    lambda x: pd.Series({
        "median": weighted_quantile(x["RIDAGEYR"], x[W], 0.5),
        "Q1": weighted_quantile(x["RIDAGEYR"], x[W], 0.25),
        "Q3": weighted_quantile(x["RIDAGEYR"], x[W], 0.75)
    })
)

age_stats["IQR"] = age_stats["Q3"] - age_stats["Q1"]

age_stats
```

```python
sns.boxplot(data=analytic, x="diagnosis_status", y="RIDAGEYR")
plt.title("Age Distribution by Diagnosis Status")
plt.show()
```

```python
glu = analytic[["LBXGLU", "diagnosis_status", W]].dropna()

glu_stats = glu.groupby("diagnosis_status").apply(
    lambda x: pd.Series({
        "median": weighted_quantile(x["LBXGLU"], x[W], 0.5),
        "Q1": weighted_quantile(x["LBXGLU"], x[W], 0.25),
        "Q3": weighted_quantile(x["LBXGLU"], x[W], 0.75)
    })
)

glu_stats["IQR"] = glu_stats["Q3"] - glu_stats["Q1"]

glu_stats
```

```python
sns.boxplot(data=analytic, x="diagnosis_status", y="LBXGLU")
plt.title("Fasting Glucose Distribution by Diagnosis Status")
plt.show()
```

```python
race_undiag = analytic[["RIDRETH1", "undiagnosed", W]].dropna()

race_undiag_prop = (
    race_undiag.groupby(["RIDRETH1", "undiagnosed"], as_index=False)[W].sum()
)

race_undiag_prop["prop"] = race_undiag_prop[W] / race_undiag_prop.groupby("RIDRETH1")[W].transform("sum")

# keep only undiagnosed rows
race_undiag_only = race_undiag_prop[race_undiag_prop["undiagnosed"] == True]

race_undiag_only
```

```python
race_undiag_only["RIDRETH1"] = race_undiag_only["RIDRETH1"].replace({
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    5: "Other"
})

sns.barplot(data=race_undiag_only, x="RIDRETH1", y="prop")

plt.ylabel("Weighted % Undiagnosed")
plt.xlabel("Race/Ethnicity")
plt.title("Weighted % Undiagnosed by Race/Ethnicity")
plt.ylim(0,1)
plt.xticks(rotation=30)

plt.show()
```

```python
gender_undiag = analytic[["RIAGENDR", "undiagnosed", W]].dropna()

gender_undiag_prop = (
    gender_undiag.groupby(["RIAGENDR", "undiagnosed"], as_index=False)[W].sum()
)

gender_undiag_prop["prop"] = gender_undiag_prop[W] / gender_undiag_prop.groupby("RIAGENDR")[W].transform("sum")

# keep only undiagnosed
gender_undiag_only = gender_undiag_prop[gender_undiag_prop["undiagnosed"] == True]

gender_undiag_only
```

```python
gender_undiag_only["RIAGENDR"] = gender_undiag_only["RIAGENDR"].replace({
    1: "Male",
    2: "Female"
})

sns.barplot(data=gender_undiag_only, x="RIAGENDR", y="prop")

plt.ylabel("Weighted % Undiagnosed")
plt.xlabel("Gender")
plt.title("Weighted % Undiagnosed by Gender")
plt.ylim(0,1)

plt.show()
```
